```python
# There are issues with the headers and the last two rows of the data
loan_db = pd.read_csv('LoanStats_2018Q4.csv', engine='python', header=1, skipfooter=2,)
loan_db.head()
```

```python
# Create function to change all values of 'term' to integers
def term_ints(terms):
  return int(terms.strip().strip(' months'))
```

```python
# check for NaN values in the column
loan_db['term'].isnull().sum()
```

```python
# apply function to column and update column
loan_db['term'] = loan_db['term'].apply(term_ints)
loan_db['term'].head()
```

```python
# make a column named loan_status_is_great. It should contain the integer 1 
# if loan_status is "Current" or "Fully Paid." 
# Else it should contain the integer 0.

loan_db['loan_status'].value_counts()
```

```python
# Create new column and set values using .loc[]
loan_db['loan_status_is_great'] = 0
loan_db.loc[loan_db['loan_status'] == 'Current', 'loan_status_is_great'] = 1
loan_db.loc[loan_db['loan_status'] == 'Fully Paid', 'loan_status_is_great'] = 1

# Checking the value counts of the new column
loan_db['loan_status_is_great'].value_counts()
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns.
loan_db['last_pymnt_d'].head()
```

```python
# make a datetime version of last_pymnt_d
loan_db['last_pymnt_dt'] = pd.to_datetime(loan_db['last_pymnt_d'].fillna(0), infer_datetime_format=True)
loan_db['last_pymnt_dt'].head()
```

```python
# Creating last_pymnt_d_month and checking if it worked...
loan_db['last_pymnt_d_month'] = loan_db['last_pymnt_dt'].dt.month
loan_db['last_pymnt_d_month'].head()
```

```python
# Creating last_pymnt_d_month and checking if it worked...
loan_db['last_pymnt_d_year'] = loan_db['last_pymnt_dt'].dt.year
loan_db['last_pymnt_d_year'].head()
```
